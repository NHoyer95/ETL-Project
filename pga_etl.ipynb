```python
# Dependencies
import pandas as pd
from sqlalchemy import create_engine
import config
```

```python
pga_2017_file = "Resources/pga_tour_stats_2017.csv"
pga_2017_df = pd.read_csv(pga_2017_file)
pga_2017_df.head()
```

```python
pga_2018_file = "Resources/pga_tour_stats_2018.csv"
pga_2018_df = pd.read_csv(pga_2018_file)
pga_2018_df.head()
```

```python
# Columns we want to extract from dataset
pga_2017_cols = ["Player", "ROUNDS_PLAYED", "POINTS", "AVG_SCORE", "NUMBER_OF_WINS", "NUMBER_OF_TOP_Tens", 
                 "AVG_CARRY_DISTANCE", "SG_PUTTING_PER_ROUND"]
pga_2017_transformed_df = pga_2017_df[pga_2017_cols].copy()

# Renaming Columns
pga_2017_transformed_df = pga_2017_transformed_df.rename(columns={"Player": "player",
                                                                  "ROUNDS_PLAYED": "rounds_played_17",
                                                                  "POINTS": "points_17", 
                                                                  "AVG_SCORE": "avg_score_17", 
                                                                  "NUMBER_OF_WINS": "wins_17", 
                                                                  "NUMBER_OF_TOP_Tens": "top_tens_17",
                                                                  "AVG_CARRY_DISTANCE": "avg_carry_distance_17",
                                                                  "SG_PUTTING_PER_ROUND": "sg_putting_per_round_17"})

# Drop data for rows that contain no data at all
pga_2017_transformed_df.dropna(how='any')

pga_2017_transformed_df.set_index("player", inplace=True)
pga_2017_transformed_df.head()
```

```python
# Columns we want to extract from dataset
pga_2018_cols = ["Player Name", "Rounds", "Points", "Average Score", "Wins", "Top 10", "Avg Distance", 
                 "Average SG Putts"]
pga_2018_transformed_df = pga_2018_df[pga_2018_cols].copy()

# Renaming Columns
pga_2018_transformed_df = pga_2018_transformed_df.rename(columns={"Player Name": "player",
                                                                  "Rounds": "rounds_played_18",
                                                                  "Points": "points_18", 
                                                                  "Average Score": "avg_score_18", 
                                                                  "Wins": "wins_18",
                                                                  "Top 10": "top_tens_18", 
                                                                  "Avg Distance": "avg_carry_distance_18",
                                                                  "Average SG Putts": "sg_putting_per_round_18"})
# Drop data for rows that contain no data at all
pga_2018_transformed_df.dropna(how='all')

pga_2018_transformed_df.set_index("player", inplace=True)
pga_2018_transformed_df.head()
```

```python
# change the data type to a float
pga_2018_transformed_df['rounds_played_18'] = pga_2018_transformed_df[['rounds_played_18']].astype(float)
print(pga_2018_transformed_df.dtypes)
```

```python
# taking out the comma's within the column to allow us to change data type
pga_2018_transformed_df['points_18'] = pga_2018_transformed_df['points_18'].str.replace(',', '') 
pga_2018_transformed_df.head()
```

```python
pga_2018_transformed_df['points_18'] = pga_2018_transformed_df[['points_18']].astype(float)
print(pga_2018_transformed_df.dtypes)
```

```python
# Create a SQL Database connection
connection_string = f"postgres:{config.password}@localhost:5432/pga_tour_stats_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# checking tables
engine.table_names()
```

```python
# Load dataframes into database
try:
    pga_2018_transformed_df.to_sql(name='pga_tour_stats_2018', con=engine, if_exists='append', index=True) 
    print("Data loaded successfully")
    
except:
    print("Data has already been loaded to db")
```

```python
try:
    pga_2017_transformed_df.to_sql(name='pga_tour_stats_2017', con=engine, if_exists='append', index=True) 
    print("Data loaded successfully")

except:
    print("Data has already been loaded to db")
```
